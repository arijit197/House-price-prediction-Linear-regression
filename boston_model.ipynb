```python
import pandas as pd
import tensorflow as tf
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.boston_housing.load_data()
y_test
```

```python
mean = tf.math.reduce_mean(x_train,axis=0)
std = tf.math.reduce_std(x_train,axis=0)
print(std)
```

```python
x_train[:,:3] = (x_train[:,:3] -mean[:3])/std[:3]
x_train[:,4:] = (x_train[:,4:] -mean[4:])/std[4:]
```

```python
x_test[:,:3] = (x_test[:,:3] -mean[:3])/std[:3]
x_test[:,4:] = (x_test[:,4:] -mean[4:])/std[4:]
```

```python
def ProbRegression():
  output = tf.keras.layers.Dense(1,input_shape=(13,))
  layers=[output]
  model=tf.keras.Sequential(layers)
  model.compile(loss='MSE', optimizer='adam',metrics='mean_squared_error')
  return model
```

```python
Model = ProbRegression()
Model.fit(x_train,y_train,epochs=10000,validation_data=(x_test,y_test))








```
